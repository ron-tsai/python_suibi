```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 
plt.rcParams['font.family'] = 'Songti SC'  # 
plt.rcParams['axes.unicode_minus'] = False  # 

# 
fig, ax = plt.subplots(figsize=(10, 6))

# 
steps = [
    {"name": "", "pos": (0.1, 0.8)},
    {"name": "", "pos": (0.1, 0.6)},
    {"name": "", "pos": (0.1, 0.4)},
    {"name": "", "pos": (0.1, 0.2)},
    {"name": "", "pos": (0.1, 0.0)},
]

# 
for step in steps:
    x, y = step["pos"]
    rect = patches.Rectangle((x, y), 0.8, 0.15, linewidth=1, edgecolor='black', facecolor='lightblue', alpha=0.8)
    ax.add_patch(rect)
    plt.text(x + 0.4, y + 0.075, step["name"], ha='center', va='center', fontsize=10)

# 
for i in range(len(steps) - 1):
    x_start = steps[i]["pos"][0] + 0.4
    y_start = steps[i]["pos"][1]
    x_end = steps[i + 1]["pos"][0] + 0.4
    y_end = steps[i + 1]["pos"][1] + 0.15
    ax.arrow(x_start, y_start, 0, y_end - y_start - 0.15, head_width=0.03, head_length=0.03, fc='black', ec='black')

# 
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')  # 

# 
plt.title("", fontsize=16, pad=20)

# 
plt.tight_layout()
plt.show()
```
