```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 设置中文字体
plt.rcParams['font.family'] = 'Songti SC'  # 使用宋体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建画布
fig, ax = plt.subplots(figsize=(10, 6))

# 定义步骤和内容
steps = [
    {"name": "文献综述与理论框架构建", "pos": (0.1, 0.8)},
    {"name": "现状调研与问题辨识", "pos": (0.1, 0.6)},
    {"name": "客观基础研究", "pos": (0.1, 0.4)},
    {"name": "影响因子分析与评价模型构建", "pos": (0.1, 0.2)},
    {"name": "创新培养模式构建与对策措施提出", "pos": (0.1, 0.0)},
]

# 绘制步骤框
for step in steps:
    x, y = step["pos"]
    rect = patches.Rectangle((x, y), 0.8, 0.15, linewidth=1, edgecolor='black', facecolor='lightblue', alpha=0.8)
    ax.add_patch(rect)
    plt.text(x + 0.4, y + 0.075, step["name"], ha='center', va='center', fontsize=10)

# 绘制箭头
for i in range(len(steps) - 1):
    x_start = steps[i]["pos"][0] + 0.4
    y_start = steps[i]["pos"][1]
    x_end = steps[i + 1]["pos"][0] + 0.4
    y_end = steps[i + 1]["pos"][1] + 0.15
    ax.arrow(x_start, y_start, 0, y_end - y_start - 0.15, head_width=0.03, head_length=0.03, fc='black', ec='black')

# 设置画布属性
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')  # 关闭坐标轴

# 添加标题
plt.title("技术路线图", fontsize=16, pad=20)

# 显示图形
plt.tight_layout()
plt.show()
```
