```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 读取Excel数据
df = pd.read_excel("词云图.xlsx", sheet_name="Sheet1")

# 将数据转换为字典（词汇: 频次）
word_freq = dict(zip(df["高频词汇"], df["频次"]))

# 配置词云参数（指定中文字体路径）
# macOS 系统中常见的中文字体路径
font_path = "/System/Library/Fonts/Supplemental/Arial Unicode.ttf"  # 修改为实际字体路径
wc = WordCloud(
    font_path=font_path,
    background_color="white",
    width=1200,
    height=800,
    max_words=100
)

# 生成词云
wordcloud = wc.generate_from_frequencies(word_freq)

# 显示词云图
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
